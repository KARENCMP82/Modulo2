**Importante**: Conversión de tipos con ndarray.type (atributo) y ndarray.astype() (método) - Diferencias:

- **ndarray.dtype = "tipo":**

  - Cambia el tipo de datos directamente en la instancia del array NumPy.
  - Modifica el tipo de datos del array inplace, es decir, modifica el propio array.
  - El argumento "tipo" debe ser una cadena de texto que represente el tipo de datos deseado, por ejemplo, "int", "float", "str", entre otros.
  - Es una forma directa de cambiar el tipo de datos del array sin crear una nueva copia.
- **ndarray.astype("tipo")**:

  - Devuelve una copia del array NumPy con el tipo de datos especificado.
  - No modifica el array original, sino que crea una nueva copia con el tipo de datos deseado.
  - El argumento "tipo" debe ser una cadena de texto que represente el tipo de datos deseado, por ejemplo, "int", "float", "str", entre otros.
  - Es útil cuando se desea mantener el array original sin cambios y crear una nueva versión con un tipo de datos diferente.

```python
# Crear un array NumPy con tipo de datos inicial 'int'
array_copia = array.copy()
```

```python
# Usar ndarray.type = "tipo" para cambiar el tipo de datos inplace
array_copia.dtype = "float64"
print("Array con tipo de datos modificado inplace:")
print(array_copia)
```

```python
# Usar ndarray.astype("tipo") para crear una nueva copia con tipo de datos modificado
array_modificado = array_copia.astype("str")
print("\nNueva copia del array con tipo de datos modificado:")
print(array_modificado)
```

```python
# Verificar el tipo de datos del array original (no ha sido modificado)
print("\nTipo de datos del array original:")
print(array_modificado.dtype)
```

## Funciones y Métodos Numpy

La biblioteca NumPy ofrece Funciones y Métodos. Es importante prestar atención a sus principales características, ya que podemos expandir sus aplicaciones a otras estructuras y bibliotecas de Python, ampliando así sus capacidades:

- **Funciones de NumPy**:

  - Se llaman directamente desde el espacio de nombres numpy.
  - Actúan sobre los arrays como argumentos de entrada y devuelven un resultado.
  -No modifican los arrays originales, sino que generan nuevos arrays o realizan operaciones sobre los arrays de entrada.
  - Ejemplos: *numpy.sum(), numpy.mean(), numpy.max(), numpy.min()*.

- **Métodos de NumPy**:

  - Se llaman en instancias de arrays NumPy específicos.
  - Operan directamente en los arrays sobre los que son llamados.
  - Pueden modificar el array original o devolver un nuevo array resultante de la operación.
 - Ejemplos: **ndarray.sum(), ndarray.mean(), ndarray.max(), ndarray.min()*.

Aunque algunas funciones y métodos pueden realizar tareas similares, se diferencian en cómo se invocan y en cómo operan en los arrays. Por ejemplo, **numpy.sum()** es una función que **se aplica a cualquier array** (incluso lisas y tuplas), mientras que ndarray.sum() es un método específico que solo se puede llamar en un ndarray existente.

**where(condición, x,y)** (!!)

Función extremadamente versátil (se utiliza incluso con DataFrames) que permite realizar operaciones condicionales en arrays NumPy. Retorna los índices donde se cumple una condición en un array.

- condicion: es la condición que queremos verificar. Si es verdadera, se devuelve el valor correspondiente de x; de lo contrario, se devuelve el valor correspondiente de y.
- x: es el valor devuelto cuando la condición es verdadera.
- y: es el valor devuelto cuando la condición es falsa.

```python
#"Usaremos el array invertido solo para evitar coincidencias de índices y elementos en el ejemplo
array_invertido = array[::-1]
array_invertido
```

```python
# Encontrar los índices donde los valores son mayores que 2
indices = np.where(array_invertido > 2)
print(indices)

# Obtener los valores donde los valores son mayores que 2 usando los índices
valores_filtrados = array_invertido[indices]
print(valores_filtrados)
```
